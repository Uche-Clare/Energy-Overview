# (Extract, Transform and Load) ETL Process and Data Observation

```python
#import libraries 
import pandas as pd
import numpy as np
```

```python
#load the data 
energy_file = 'World Energy Consumption.csv'
df_energy = pd.read_csv(energy_file)
df_energyc=df_energy.copy()
```

```python
#display all columns
pd.set_option('display.max_columns', None)
df_energyc
```

```python
#get info about the data
df_energyc.info()
```

```python
#determine the number of rows and columns
df_energyc.shape
```

```python
#Analyze this data set with 'describe'
df_energyc.describe()
```

```python
#The year column is seen as a float datatype, we need to convert it to datetime.Convert the year column to datetime
df_energyc['year']=pd.to_datetime(df_energyc['year'], format='%Y')
```

# Check for Missing and Duplicates Values

```python
#determine the number of null values in each column
null_colns = df_energyc.isna().sum()
print(null_colns)
```

```python
#Check for duplicate values 
df_energyc.duplicated().sum()
```

# Extract Data for 3 african countries: Nigeria, Rwanda. Ghana

```python
#display 3 african countries 
country_to_display = ['Nigeria', 'Rwanda', 'Ghana']
df_energyc = df_energyc[df_energyc['country'].isin(country_to_display)]
```

```python
#what is the data shape/dimension of the data for these countries?
df_energyc.shape
```

```python
#create a copy of the new dataset
df_energy_afr=df_energyc.copy()
```

```python
#determine the energy generation and consumption of these countries the past 20 years. So we will filter from 2000 to 2020
start_year = 2000
end_year = 2020 
df_energy_afr = df_energy_afr[(df_energy_afr['year'].dt.year >= start_year) & (df_energy_afr['year'].dt.year <= end_year)]
```

```python
#display the data
df_energy_afr
```

# Data Cleaning: Handling NAN and Zero columns

```python
#notice that they're a lot of empty column. We'll dropping all empty columns
df_energy_afr.dropna(axis=1, how='all', inplace = True)
```

```python
#display data
df_energy_afr
```

```python
# Identify columns that contain all zeros
NaN_columns = df_energy_afr.columns[df_energy_afr.isna().all()]
NaN_columns
```

```python
#we'll be dropping these columns since there's no data for them
df_energy_afr.drop(NaN_columns, axis=1, inplace=True)
```

```python
#display data
df_energy_afr
```

```python
# display only selected columns
selected_columns = ["iso_code", "country", "year", "electricity_generation", "fossil_electricity", 
                    "hydro_electricity","renewables_electricity", "solar_electricity","energy_per_capita",
                    "fossil_cons_per_capita", "fossil_share_elec", "hydro_share_elec", "hydro_elec_per_capita",
                    "population","solar_share_elec","solar_elec_per_capita","gdp"]
df_energy_afr[selected_columns]
df_en_afr= df_energy_afr[selected_columns].copy()
```

```python
#check for null values in each column
null_colns = df_en_afr.isna().sum()
print(null_colns)
# as seen below columns like energy_per_capita,population and gdp still have empty values which can lead to data inconsistency 
```

```python
#replace every null value in energy_per_capita with interpolation 
df_en_afr['energy_per_capita'].interpolate(inplace=True)
```

```python
#replace every null value in populatiion with interpolation 
df_en_afr['population'].interpolate(inplace=True)
```

```python
#replace every null value in gdp with interpolation 
df_en_afr['gdp'].interpolate(inplace=True)
df_en_afr
```

```python
#display all values in the row
pd.set_option('display.max_rows', None)
df_en_afr
```
